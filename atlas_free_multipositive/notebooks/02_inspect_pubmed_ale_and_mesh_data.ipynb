```python
from pathlib import Path
import json, pandas as pd, torch, yaml

cfg_path = Path('atlas_free_multipositive/configs/paths.yaml')
paths_cfg = yaml.safe_load(cfg_path.read_text())
paths = {
    'ale_cache': paths_cfg['existing_ale_cache'],
    'mesh': paths_cfg['pubmed_mesh_annotations'],
    'nodes': paths_cfg['pubmed_mesh_nodes'],
    'summaries': paths_cfg['pubmed_summaries'],
}
for k, p in paths.items():
    print(k, Path(p).exists(), p)
missing = {k: p for k, p in paths.items() if not Path(p).exists()}
if missing:
    raise FileNotFoundError(f'Missing configured files: {missing}. Update atlas_free_multipositive/configs/paths.yaml to match where you moved the data.')

cache = torch.load(paths['ale_cache'], map_location='cpu', weights_only=False)
print(cache['volumes'].shape, len(cache['pmids']))
nodes = pd.read_parquet(paths['nodes'])
print(nodes.shape, nodes.columns.tolist())
display(nodes.head())
```
